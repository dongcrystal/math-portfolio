```python
# 벡터를 리스트로 표현
v1 = [3, 2]
v2 = [1, 4]
```

```python
print(v1)
print(v2)
```

```python
# 벡터 덧셈 함수
def vector_add(a, b):
    return [a[i] + b[i] for i in range(len(a))]
```

```python
print(vector_add(v1, v2))
```

```python
# 스칼라 곱 함수
def scalar_multiply(scalar, vector):
    return [scalar * x for x in vector]

print("v1:", v1)
print("v2:", v2)
```

```python
# 벡터 덧셈 결과
print("v1 + v2 =", vector_add(v1, v2))
```

```python
# 스칼라 3 곱하기 v1
print("3 * v1 =", scalar_multiply(3, v1))
```

```python
import matplotlib.pyplot as plt

# 벡터 정의
v1 = [3, 2]
v2 = [1, 4]

# 벡터 v1 그리기 (빨간색)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')

# 벡터 v2 그리기 (파란색)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# 그래프 꾸미기
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.title('Vectors v1 and v2')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 벡터 정의
v1 = [3, 2]
v2 = [1, 4]

# 벡터 덧셈 함수
def vector_add(a, b):
    return [a[i] + b[i] for i in range(len(a))]

# 벡터 덧셈 결과 계산
v3 = vector_add(v1, v2)

# 벡터 v1 그리기 (빨간색)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')

# 벡터 v2 그리기 (파란색)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# 벡터 덧셈 결과 v3 그리기 (초록색)
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1 + v2')

# 그래프 꾸미기
plt.xlim(-1, 6)
plt.ylim(-1, 7)
plt.grid()
plt.legend()
plt.title('Vector Addition Visualization')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 벡터 정의
v1 = [3, 2]
v2 = [1, 4]

# 벡터 덧셈 함수
def vector_add(a, b):
    return [a[i] + b[i] for i in range(len(a))]

# 스칼라 곱 함수
def scalar_multiply(scalar, vector):
    return [scalar * x for x in vector]

# 벡터 덧셈 결과 계산
v3 = vector_add(v1, v2)

# 스칼라 곱 결과 계산 (3 * v1)
v4 = scalar_multiply(3, v1)

# 벡터 v1 그리기 (빨간색)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')

# 벡터 v2 그리기 (파란색)
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# 벡터 덧셈 결과 v3 그리기 (초록색)
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1 + v2')

# 스칼라 곱 결과 v4 그리기 (보라색)
plt.quiver(0, 0, v4[0], v4[1], angles='xy', scale_units='xy', scale=1, color='m', label='3 * v1')

# 그래프 꾸미기
plt.xlim(-1, 12)
plt.ylim(-1, 12)
plt.grid()
plt.legend()
plt.title('Vector Addition and Scalar Multiplication Visualization')
plt.show()
```
